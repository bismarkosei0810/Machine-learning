```python
import pandas as pd
```

### Create a new variable called file_url that will contain the URL to theraw dataset:

```python
file_url = (
    'https://raw.githubusercontent.com/'
    'PacktWorkshops/'
    'The-Applied-Artificial-Intelligence-Workshop/'
    'master/Datasets/car.csv'
)
```

### Load the data using the pd.read_csv() method.:

```python
df = pd.read_csv(file_url)
```

```python
from sklearn import preprocessing
def encode(data_frame, column):
    label_encoder = preprocessing.LabelEncoder()
    label_encoder.fit(data_frame[column].unique())
    return label_encoder.transform(data_frame[column])
for column in df.columns:
    df[column] = encode(df, column)
label = df.pop('class')
```

```python
from sklearn import model_selection
features_train, features_test, label_train, label_test = model_selection.train_test_split(df, label, test_size=0.1, random_state=88)
```

### Import RandomForestClassifier from sklearn.ensemble:

```python
from sklearn.ensemble import RandomForestClassifier
```

### instantiate a random forest classifier with n_estimators=100,max_depth=6, and random_state=168. Save it to a variable calledrandom_forest_classifier:

```python
random_forest_classifier = RandomForestClassifier(n_estimators=100, max_depth=6, random_state=168)
```

### Fit the random forest classifier with the training set:

```python
random_forest_classifier.fit(features_train, label_train)
```

### Make predictions on the testing set using the random forest classifier and save them in a variable called rf_preds_test. Print its content:

```python
rf_preds_test = random_forest_classifier.fit(features_train, label_train)
rf_preds_test
```

### import classification_report from sklearn.metrics:

```python
from sklearn.metrics import classification_report
```

### Print the classification report with the labels and predictions from the test set:

```python
type(rf_preds_test)
```

```python
rf_preds_test = random_forest_classifier.predict(features_test)
```

```python
print(classification_report(label_test, rf_preds_test))
```

###  In the above result,the model executed successfully without any errors. All preprocessing, training, and evaluation steps ran as expected, and the results generated reflect the model’s current performance. The workflow is stable and ready for further analysis or tuning,.
### The F1 score in the preceding report shows us that the random forest is performing well on class 2 but not as good for classes 0 and 3. The model is unable to predict accurately for class 1, but there were only 9 observations in the testing set. The accuracy score is 0.84, while the F1 score is 0.82.

```python
from sklearn.metrics import confusion_matrix
```

### Display the confusion matrix on the true and predicted labels of the testing set:

```python
confusion_matrix(label_test, rf_preds_test)
```

### From this confusion matrix, we can see that the RandomForest model ishaving difficulties accurately predicting the first class. It incorrectly predicted 16 cases (8 + 5 + 3) for this class.
### Print the feature importance score of the test set using .feature_ importance_ and save the results in a variable called rf_varimp. Print its contents:

```python
rf_varimp = random_forest_classifier.feature_importances_
rf_varimp
```
